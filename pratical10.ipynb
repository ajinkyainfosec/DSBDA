```python
Q1 = df['sepal_width'].quantile(0.25)
Q3 = df['sepal_width'].quantile(0.75)
IQR = Q3 - Q1
outliers = df[(df['sepal_width'] < (Q1 - 1.5 * IQR)) | (df['sepal_width'] > (Q3 + 1.5 * IQR))]
```

```python
outliers
```

```python
sns.boxplot(x = 'sepal_width' , data = outliers)
```
